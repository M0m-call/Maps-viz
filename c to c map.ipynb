```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

```python
R = np.linspace(0,2,60)
THETA = np.linspace(0, np.pi*2, 120)
radii, thetas = np.meshgrid(R,THETA)
```

```python
def f(z):
    return np.exp(z)
```

```python
nprect = np.vectorize(cmath.rect)
z = nprect(radii,thetas) 
```

```python
#Centre of our ball
z =  z + complex(0,0)
```

```python
#input space
size = [-3,3]
#output sapce
size2 = [-3,3]
```

```python

#ax1.plot(thetas, radii, marker='.', ls='none')
#ax2.plot(thetas, radii, marker='.', ls='none')
ax1 = plt.subplot(121)
ax1.plot(z.real, z.imag, marker='.', ls='none')
plt.title("Input")
plt.axvline(x=0, label="Imaginary")
plt.axhline(y=0, label="Real")
plt.xlim(size)
plt.ylim(size)
ax2 = plt.subplot(122)
ax2.plot(f(z).real ,f(z).imag, marker='.', ls='none')
#ax1.lab
plt.title("Output")
plt.axvline(x=0,  label="Imaginary")
plt.axhline(y=0,  label="Real")
plt.xlim(size2)
plt.ylim(size2)
plt.show()
```

```python
def diam(arr):
    a,b=np.max(arr),np.min(arr)
    return np.abs(a-b)
```

```python
def slope(a,b):
    
    m= (b.imag - a.imag)/(b.real-a.real)
    
    return m
    
```

```python
print(f'diam of ball : {diam(z)}')
print(f'diam after applying f : {diam(f(z))}')
ratio = diam(f(z))/diam(z)
```

```python
print(f"ratio : {ratio} an approximation of |f'(z)|  ") 
print(f"actual |f'(z)| = 1")
```
